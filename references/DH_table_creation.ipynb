# DH Table Creation Guide ‚Äî Modified DH Convention

**Source:** CMU 16-311 Lecture 16b
**Reference:** https://www.cs.cmu.edu/~16311/f05/lecture/lec16b.html

---

## Key Difference from Standard DH

In standard DH, $z_i$ is the axis of joint $i+1$.
In **modified DH**, $z_i$ is the axis of joint $i$ ‚Äî the indexing matches directly.

This also changes how $x_i$ is assigned and how the parameters are indexed.

---

## Step 1 ‚Äî Label the joint axes $z_1, \ldots, z_n$

Assign $z_i$ along the axis of **joint $i$**:
- **Revolute joint:** $z_i$ is the axis of rotation
- **Prismatic joint:** $z_i$ is the direction of extension

---

## Step 2 ‚Äî Establish the base frame $\{0\}$

- Place origin $O_0$ anywhere convenient
- Choose $x_0$ pointing along the common perpendicular from $z_0$ toward $z_1$
- Choose $y_0$ to complete a right-handed frame

---

## Steps 3‚Äì4 ‚Äî Assign frames $\{1\}$ through $\{n\}$

Repeat for each joint $i = 1, \ldots, n$:

**Step 3 ‚Äî Locate origin $O_i$:**

| Relationship between $z_i$ and $z_{i+1}$ | Place $O_i$ at... |
|---|---|
| Skew (general case) | where the common perpendicular to $z_i$ and $z_{i+1}$ intersects $z_i$ |
| Intersecting | the intersection point |
| Parallel | any convenient point along $z_i$ |

**Step 4 ‚Äî Establish $x_i$:**
- Points along the common perpendicular from $z_i$ toward $z_{i+1}$
- If $z_i$ and $z_{i+1}$ intersect: $x_i$ points perpendicular to the plane containing $z_i$ and $z_{i+1}$

**Step 5 ‚Äî Establish $y_i$** to complete a right-handed frame. (Rarely drawn explicitly.)

---

## Step 6 ‚Äî Establish the end-effector frame $\{n\}$

- Place $O_n$ at the tool center point
- $z_n$ along the approach direction
- $x_n$ and $y_n$ chosen conveniently to form a right-handed frame

---

## Step 7 ‚Äî Fill out the DH parameter table

Modified DH uses **mixed indexing** ‚Äî the link parameters carry subscript $i-1$ and the joint parameters carry subscript $i$:

| Parameter | Definition | Variable? |
|---|---|---|
| $a_{i-1}$ | Distance along $x_{i-1}$ between $z_{i-1}$ and $z_i$ | No ‚Äî fixed link geometry |
| $\alpha_{i-1}$ | Angle from $z_{i-1}$ to $z_i$, measured about $x_{i-1}$ (right-hand rule) | No ‚Äî fixed link geometry |
| $d_i$ | Distance along $z_i$ from $x_{i-1}$ to $x_i$ | **Yes, if joint $i$ is prismatic** |
| $\theta_i$ | Angle from $x_{i-1}$ to $x_i$, measured about $z_i$ (right-hand rule) | **Yes, if joint $i$ is revolute** |

> **Revolute joint:** $\theta_i$ is the variable. Set it to $0$ at home position. $d_i$ is a fixed constant.
>
> **Prismatic joint:** $d_i$ is the variable. Set it to $0$ at home position. $\theta_i$ is a fixed constant.
>
> $a_{i-1}$ and $\alpha_{i-1}$ are always fixed ‚Äî they describe the physical geometry of the link.

> **Units:** distances in metres (or a consistent unit), angles in **radians**.
> Common values: $90¬∞ = 1.5708\ \text{rad}$, $180¬∞ = 3.1416\ \text{rad}$

---

## Quick Reference ‚Äî Special Cases

| Relationship between $z_i$ and $z_{i+1}$ | Consequence |
|---|---|
| Parallel | $\alpha_{i-1} = 0$, choose $O_i$ to simplify (often sets $d_i = 0$) |
| Intersecting | $a_{i-1} = 0$, place $O_i$ at the intersection |
| Skew | General case ‚Äî unique common perpendicular |
