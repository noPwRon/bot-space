# DH Table Creation Guide
**Convention: Illinois Lab 3 Guide (based on Spong)**
Reference: [Illinois Lab 3 Guide](https://walt.web.illinois.edu/wp-content/uploads/2021/11/470-Lab-3-DH-Frame-Fwd-Kin-Explanation-Spring-2019.pdf)

> All four parameters in row $i$ share the same subscript $i$. There is no mixed indexing.

---

## Step 1 — Label the joint axes $z_0, z_1, \ldots, z_{n-1}$

Each $z_i$ is the axis of motion for the **next** joint:
- **Revolute joint**: $z_i$ is the axis of rotation (right-hand rule sets positive direction)
- **Prismatic joint**: $z_i$ is the direction of extension

You have $n$ joints, so you will have $n$ z-axes labeled $z_0$ through $z_{n-1}$.

> Note: $z_0$ belongs to the base frame, not joint 1. So $z_0$ is the axis that joint 1 moves about or along.

---

## Step 2 — Establish the base frame $\{0\}$

- Place the origin $O_0$ anywhere on the $z_0$ axis
- Choose $x_0$ in any convenient direction perpendicular to $z_0$
- $y_0$ is determined by the right-hand rule (you rarely need to draw it)

---

## Step 3 — Locate the origin $O_i$ for each frame

For each frame $\{i\}$, find the **common normal** between $z_i$ and $z_{i-1}$, then:

| Relationship between $z_{i-1}$ and $z_i$ | Place $O_i$ at... |
|---|---|
| Skew (general case) | where the common normal intersects $z_i$ |
| Intersecting | the intersection point |
| Parallel | any convenient point along $z_i$ |

---

## Step 4 — Establish $x_i$ for each frame

- $x_i$ points along the common normal between $z_i$ and $z_{i-1}$, through $O_i$
- If $z_i$ and $z_{i-1}$ **intersect**: $x_i$ points perpendicular to the plane containing $z_{i-1}$ and $z_i$

---

## Step 5 — Establish the end-effector frame $\{n\}$

- Make $z_n$ parallel to $z_{n-1}$
- $x_n$ must intersect $z_{n-1}$
- Place $O_n$ at the tool center point (TCP)

---

## Step 6 — Fill out the DH table

The table has one row per joint. All four parameters in row $i$ share subscript $i$:

| Parameter | Definition | Variable? |
|---|---|---|
| $a_i$ | Distance along $x_i$ from where $x_i$ meets $z_{i-1}$, to $O_i$ | No |
| $\alpha_i$ | Angle from $z_{i-1}$ to $z_i$, measured about $x_i$ (right-hand rule) | No |
| $d_i$ | Distance along $z_{i-1}$ from $O_{i-1}$ to where $x_i$ meets $z_{i-1}$ | **Yes, if prismatic** |
| $\theta_i$ | Angle from $x_{i-1}$ to $x_i$, measured about $z_{i-1}$ (right-hand rule) | **Yes, if revolute** |

> For a revolute joint: $\theta_i$ is variable, set it to $0$ at the home position (or the offset value).
> For a prismatic joint: $d_i$ is variable, set it to $0$ at the home position (or the offset value).
> All other values are fixed constants describing the robot's physical geometry.

> **Units**: all distances in metres (or a consistent unit), all angles in **radians**.
> Common values: $90° = 1.5708\ \text{rad}$, $180° = 3.1416\ \text{rad}$

---

## Step 7 — Build the transformation matrix for each joint

Each row of the DH table produces one $4 \times 4$ homogeneous transform $A_i$:

$$A_i = \text{Rot}_{z,\theta_i}\ \text{Trans}_{z,d_i}\ \text{Trans}_{x,a_i}\ \text{Rot}_{x,\alpha_i}$$

$$A_i = \begin{bmatrix} \cos\theta_i & -\sin\theta_i\cos\alpha_i & \sin\theta_i\sin\alpha_i & a_i\cos\theta_i \\ \sin\theta_i & \cos\theta_i\cos\alpha_i & -\cos\theta_i\sin\alpha_i & a_i\sin\theta_i \\ 0 & \sin\alpha_i & \cos\alpha_i & d_i \\ 0 & 0 & 0 & 1 \end{bmatrix}$$

The full forward kinematics from base to end-effector is:

$$T_n^0 = A_1 \cdot A_2 \cdot A_3 \cdots A_n$$

The top-left $3 \times 3$ block is the rotation matrix. The top-right $3 \times 1$ column is the position of the end-effector in the base frame.
