# DH Table Creation Guide

**Source:** Spong, M.W., Hutchinson, S., Vidyasagar, M. — *Robot Modeling and Control*, Chapter 3
**Reference PDF:** https://users.cs.duke.edu/~brd/Teaching/Bio/asmb/current/Papers/chap3-forward-kinematics.pdf

---

## Step 1 — Label the joint axes $z_0, \ldots, z_{n-1}$

$z_i$ is the axis of actuation for joint $i+1$. So $z_0$ is the axis of joint 1, $z_1$ is the axis of joint 2, and so on.

- **Revolute joint:** $z_i$ is the axis of rotation
- **Prismatic joint:** $z_i$ is the direction of extension

You have $n$ joints, so you will label $n$ z-axes: $z_0$ through $z_{n-1}$.

---

## Step 2 — Establish the base frame

- Place origin $O_0$ anywhere on $z_0$
- Choose $x_0$ and $y_0$ to form a right-handed frame
- The choice is arbitrary — pick whatever makes the geometry simplest

---

## Steps 3–5 — Assign frames $\{1\}$ through $\{n-1\}$

Repeat for each $i = 1, \ldots, n-1$:

**Step 3 — Locate origin $O_i$:**

| Relationship between $z_{i-1}$ and $z_i$ | Place $O_i$ at... |
|---|---|
| Skew (general case) | where the common normal to $z_{i-1}$ and $z_i$ intersects $z_i$ |
| Intersecting | the intersection point |
| Parallel | any convenient point along $z_i$ |

**Step 4 — Establish $x_i$:**
- Points along the common normal between $z_{i-1}$ and $z_i$, through $O_i$
- If $z_{i-1}$ and $z_i$ intersect: $x_i$ points normal to the plane containing $z_{i-1}$ and $z_i$

**Step 5 — Establish $y_i$** to complete a right-handed frame. (Rarely needed explicitly.)

---

## Step 6 — Establish the end-effector frame

- Place $O_n$ at the gripper center or tool tip
- $z_n$ points along the approach direction (along $z_{n-1}$)
- $y_n$ points in the gripper closure direction
- $x_n = y_n \times z_n$
- If no gripper, choose $x_n$ and $y_n$ conveniently

---

## Step 7 — Fill out the DH parameter table

One row per joint. All four parameters share subscript $i$:

| Parameter | Definition | Variable? |
|---|---|---|
| $a_i$ | Distance along $x_i$ from $O_i$ to the intersection of $x_i$ and $z_{i-1}$ | No — fixed link geometry |
| $d_i$ | Distance along $z_{i-1}$ from $O_{i-1}$ to the intersection of $x_i$ and $z_{i-1}$ | **Yes, if joint $i$ is prismatic** |
| $\alpha_i$ | Angle from $z_{i-1}$ to $z_i$, measured about $x_i$ (right-hand rule) | No — fixed link geometry |
| $\theta_i$ | Angle from $x_{i-1}$ to $x_i$, measured about $z_{i-1}$ (right-hand rule) | **Yes, if joint $i$ is revolute** |

> **Revolute joint:** $\theta_i$ is the variable. Set it to $0$ at home position (or the offset if one exists). $d_i$ is a fixed constant.
>
> **Prismatic joint:** $d_i$ is the variable. Set it to $0$ at home position (or the offset if one exists). $\theta_i$ is a fixed constant.
>
> $a_i$ and $\alpha_i$ always describe fixed physical geometry — they never change.

> **Units:** distances in metres (or a consistent unit), angles in **radians**.
> Common values: $90° = 1.5708\ \text{rad}$, $180° = 3.1416\ \text{rad}$

---

## Quick Reference — Special Cases

| Relationship between $z_{i-1}$ and $z_i$ | Consequence |
|---|---|
| Parallel | $\alpha_i = 0$, choose $O_i$ to simplify (often sets $d_i = 0$) |
| Intersecting | $a_i = 0$, place $O_i$ at the intersection |
| Skew | General case — unique common normal |
